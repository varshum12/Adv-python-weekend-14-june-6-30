### Aggregate functions
- sum
- max
- min
- mean
- median
- var
- std

```python
# Display total sales
df['Sales'].sum()
```

```python
round(df['Sales'].sum() , 4)
```

```python
a = df['Sales'].agg({'Sales' :  'sum'})
```

```python
round(a  , 4)
```

```python
# Maximum sales
df['Sales'].max()
```

```python
df['Sales'].agg({'Sales' : 'max'})
```

```python
# minimum sales
df['Sales'].agg({'Sales' : 'min'})
df.Sales.min()
df['Sales'].min()
```

```python
# average sale
df['Sales'].agg({'Sales' : 'mean'})
df.Sales.mean()
df['Sales'].mean()
```

```python
# median sales
df['Sales'].agg({'Sales' : 'median'})
df.Sales.median()
df['Sales'].median()
```
